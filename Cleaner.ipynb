## Clear merged branches

In this step, we remove branches that no longer exist in origin due to the merge process.

```python
! pwsh -c "git fetch --prune; git branch -vv | Where-Object { $_ -match '\[.*: gone\]' } | ForEach-Object { git branch -D (($_ -split '\s+')[1]) }"
```

## Remove build outputs

```python
! pwsh -c "Get-ChildItem -Path . -Directory -Recurse | Where-Object {($_.FullName -match '\\bin(\\x64)?\\(Debug|Release)$') -and    ($_.FullName -notmatch '\\(dbUpdater|ChapyarManager|cnprWaiting|ChapyarInitSqlInstallation|chapyarAdmin|PKCS11Admin|Chapyar.ResourceManager)(\\|$)?') } |ForEach-Object {    Get-ChildItem -Path $_.FullName -Recurse -File | Remove-Item -Force}"
```
